–ü–æ—á–Ω–µ–º–æ. –û—Å—å –≤–∞—à–µ –ø–µ—Ä—à–µ –∑–∞–≤–¥–∞–Ω–Ω—è: —Å—Ç–≤–æ—Ä—ñ—Ç—å –∑–º—ñ–Ω–Ω—É `v1`, —è–∫–∞ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—Ç–∏–º–µ –∑ —Å–µ–±–µ **–≤–µ–∫—Ç–æ—Ä-—Ä—è–¥–æ–∫** –∑ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–∏—Ö —á–∏—Å–µ–ª –≤—ñ–¥ 1 –¥–æ 5 (–≤–∫–ª—é—á–Ω–æ). –î–ª—è —Ü—å–æ–≥–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü—ñ—é arange, –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ –ø–µ—Ä–µ–≥–ª—è–Ω—É–≤—à–∏ —ó—ó –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü—ñ—é –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `?np.arange`

–¢–∞–∫–æ–∂ —Å—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é `A` —Ä–æ–∑–º—ñ—Ä—É 5x7 [*], –∑–∞–ø–æ–≤–Ω–µ–Ω—É –æ–¥–∏–Ω–∏—Ü—è–º–∏. –ö—Ä—É—Ç–æ, —è–∫—â–æ –∑–Ω–∞—î—Ç–µ –¥–µ–∫—ñ–ª—å–∫–∞ –≤–∞—Ä—ñ–∞–Ω—Ç—ñ–≤, —è–∫ —Ü–µ –∑—Ä–æ–±–∏—Ç–∏, –º–æ–∂–µ—Ç–µ –ø–æ–∫–∞–∑–∞—Ç–∏ —ó—Ö —É—Å—ñ :)

–©–æ–± –±—É–ª–æ –Ω–µ —Ç–∞–∫ –Ω—É–¥–Ω–æ –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –∑ —Ü—ñ—î—é –º–∞—Ç—Ä–∏—Ü–µ—é, –¥–∞–≤–∞–π—Ç–µ –æ–¥—Ä–∞–∑—É –∑–∞–º—ñ–Ω–∏–º–æ —É—Å—ñ –æ–¥–∏–Ω–∏—Ü—ñ –Ω–∞ –ø–µ—Ä–µ—Ç–∏–Ω—ñ —Å—Ç–æ–≤–ø—á–∏–∫—ñ–≤ 2-5 —Ç–∞ —Ä—è–¥–∫—ñ–≤ 1-4 –Ω–∞ **—á–µ—Ç–≤—ñ—Ä–∫–∏**. –Ü–Ω—à–∏–º–∏ —Å–ª–æ–≤–∞–º–∏, —è —Ö–æ—á—É —â–æ–± –∑–º—ñ–Ω–Ω–∞ `A` –º—ñ—Å—Ç–∏–ª–∞ –Ω–∞—Å—Ç—É–ø—É –º–∞—Ç—Ä–∏—Ü—é:   

$$

–ù–µ –∑–∞–±—É–¥—å—Ç–µ –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏, —â–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å–∞–º–µ —Ç–∞–∫–∏–π, —è–∫ —Ö–æ—á–µ—Ç—å—Å—è.

````python
A[0:4 , 1:5] = 4

A
````

–ù–∞ —Ü—å–æ–º—É –ø–æ–∫–∏ –≤—Å–µ, –∑—É—Å—Ç—Ä—ñ–Ω–µ–º–æ—Å—å –≤ –∞—É–¥–∏—Ç–æ—Ä—ñ—ó üòâ 

### –ö–Ü–ù–ï–¶–¨ –ü–Ü–î–ì–û–¢–û–í–ß–û–á –ß–ê–°–¢–ò–ù–ò

### –ü–û–ß–ê–¢–û–ö –°–ü–Ü–õ–¨–ù–û–á –ß–ê–°–¢–ò–ù–ò

–î–æ–¥–∞—Ç–∫–æ–≤–æ –¥–æ–¥–∞–º–æ –æ–¥–∏–Ω–∏—á–∫—É –¥–æ –ø º—è—Ç–∏ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ - –¥–æ —Ç—Ä–µ—Ç—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ —É –ø–µ—Ä—à–æ–º—É —Ä—è–¥–∫—É, —á–µ—Ç–≤–µ—Ä—Ç–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ —É –¥—Ä—É–≥–æ–º—É —Ä—è–¥–∫—É, –ø º—è—Ç–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ —É —Ç—Ä–µ—Ç—å–æ–º—É —Ä—è–¥–∫—É —ñ —Ç–¥. –Ü–Ω—à–∏–º–∏ —Å–ª–æ–≤–∞–º–∏, —Ö–æ—á–µ–º–æ –æ—Ç—Ä–∏–º–∞—Ç–∏ –Ω–∞—Å—Ç—É–ø–Ω—É –º–∞—Ç—Ä–∏—Ü—é: 

$$
\begin{pmatrix}
1 & 4 & 5 & 4 & 4 & 1 & 1 \\
1 & 4 & 4 & 5 & 4 & 1 & 1 \\
1 & 4 & 4 & 4 & 5 & 1 & 1 \\
1 & 4 & 4 & 4 & 4 & 2 & 1 \\
1 & 1 & 1 & 1 & 1 & 1 & 2 \\
\end{pmatrix}
$$
–û–±–æ–≤ º—è–∑–∫–æ–≤–æ –≤–ø–µ–≤–Ω—ñ—Ç—å—Å—è, —â–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ç–æ–π, —â–æ –≤–∏ —Ö–æ—á–µ—Ç–µ, —ñ —â–æ –≤–∏ –Ω–µ –∑—Ä–æ–±–∏–ª–∏ –æ–¥–Ω—É –π —Ç—É —Å–∞–º—É –¥—ñ—é –¥–µ–∫—ñ–ª—å–∫–∞ —Ä–∞–∑—ñ–≤!

````python
A[0,2] +=1
A[1,3] +=1
A[2,4] +=1
A[3,5] +=1
A[4,6] +=1
A
````

–¢–µ–ø–µ—Ä —Ç—Ä–æ—Ö–∏ –ø–æ–≥—Ä–∞—î–º–æ—Å—è –∑ `A` —Ç–∞ `b`. –û—Ç—Ä–∏–º–∞–π—Ç–µ –º–∞—Ç—Ä–∏—Ü—é $C$ —Ä–æ–∑–º—ñ—Ä–æ–º, –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç —è–∫–æ—ó –Ω–∞ 3 –º–µ–Ω—à–∏–π, –Ω—ñ–∂ –ø–æ–¥–≤–æ—î–Ω–∏–π –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –º–∞—Ç—Ä–∏—Ü—ñ $A$.

````python
C = (2 * A) - 3
C
````

–¢–µ–ø–µ—Ä —è —Ö–æ—á—É –æ—Ç—Ä–∏–º–∞—Ç–∏ –≤–µ–∫—Ç–æ—Ä `d`, —è–∫–∏–π —î —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º –º–Ω–æ–∂–µ–Ω–Ω—è –≤–µ–∫—Ç–æ—Ä–∞ $b$ –Ω–∞ –º–∞—Ç—Ä–∏—Ü—é $C$. –¢—É—Ç –æ–¥—Ä–∞–∑—É –ø–∏—Ç–∞–Ω–Ω—è - —É —Ü—å–æ–º—É –≤–∏–ø–∞–¥–∫—É –≤–µ–∫—Ç–æ—Ä –Ω–∞ –º–∞—Ç—Ä–∏—Ü—é –¥–æ–≤–µ–¥–µ—Ç—å—Å—è –º–Ω–æ–∂–∏—Ç–∏ –∑–ª—ñ–≤–∞ —á–∏ —Å–ø—Ä–∞–≤–∞? 

–£ –±—É–¥—å-—è–∫–æ–º—É —Ä–∞–∑—ñ, –ø—ñ–¥–∫–∞–∑–∫–∞ - —Å–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ—Å—å –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–º `@`.

````python
b = v1

d = b @ C
d
````

–¢–µ–ø–µ—Ä –ø–æ–¥–∏–≤—ñ—Ç—å—Å—è –Ω–∞ —Å–∫–∞–ª—è—Ä–Ω–∏–π –¥–æ–±—É—Ç–æ–∫ –≤–µ–∫—Ç–æ—Ä–∞ `d` —Å–∞–º–æ–≥–æ –Ω–∞ —Å–µ–±–µ. –ú–∞—î –≤–∏–π—Ç–∏ 9535 :)  

````python
sc = d @ d
sc
````

### –ú–∞—Ç–µ–º–∞—Ç–∏—á–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó

–û–∫—Ä—ñ–º –æ–ø–µ—Ä–∞—Ç–æ—Ä—ñ–≤, numpy –Ω–∞–¥–∞—î –Ω–∞–º –∫—É–ø—É —Ä—ñ–∑–Ω–∏—Ö —Ñ—É–Ω–∫—Ü—ñ–π –¥–ª—è –≤–∑–∞—î–º–æ–¥—ñ—ó –∑ n-–≤–∏–º—ñ—Ä–Ω–∏–º–∏ –º–∞—Å–∏–≤–∞–º–∏. `np.sin`, `np.exp`, `np.pow`, ... - –∫–æ—Ä–æ—Ç—à–µ, –≤—Å–µ, —á–æ–≥–æ –¥—É—à–∞ –∑–∞–±–∞–∂–∞—î. 

–ó–Ω–∞—é—á–∏ —Ü–µ–π —Ñ–∞–∫—Ç, –º–æ–∂–ª–∏–≤–æ –∑–¥–æ–≥–∞–¥–∞—î—Ç–µ—Å—å, —è–∫ –ø–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ 
$$ \sum_{i=1}^{2'000'000} e^{\sin i \cdot \cos 1/i} $$ 

````python
%%timeit
import numpy as np

i = np.arange(1, 2000001)
res = np.sum(np.pow(np.e, (np.sin(i) * np.cos(1 / i))))
res
````

–ú–∞—î –≤–∏–π—Ç–∏ —â–æ—Å—å –±–ª–∏–∑—å–∫–µ –¥–æ $2532130.799$. –°–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ—Å—å `%%timeit`, —â–æ–± –ø–æ–≥–ª—è–Ω—É—Ç–∏, —Å–∫—ñ–ª—å–∫–∏ –∑–∞–π–Ω—è–ª–∏ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è, —è–∫—ñ –≤–∏ –Ω–∞–ø–∏—Å–∞–ª–∏.

–¶—ñ–∫–∞–≤–æ, –∞ —Å–∫—ñ–ª—å–∫–∏ –± –∑–∞–π–Ω—è–ª–æ —Ü–µ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è, —è–∫–±–∏ –≤–∏ –∑–∞–º—ñ—Å—Ç—å numpy –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–ª–∏ –∑–≤–∏—á–∞–π–Ω–∏–π python? –©–æ –∂, –∑–∞–ø—É—Å—Ç—ñ—Ç—å –Ω–∞—Å—Ç—É–ø–Ω—ñ –¥–≤—ñ –∫–æ–º—ñ—Ä–∫–∏, —â–æ–± –ø–æ–≥–ª—è–Ω—É—Ç–∏:

````python
from math import sin as m_sin, cos as m_cos, exp as m_exp
````

````python
%%timeit 

result = 0
for i in range(1, 2*1000*1000 + 1):
    result += m_exp(m_sin(i) * m_cos(1 / i))
````

–î–æ —Ä–µ—á—ñ, –º–æ–∂–ª–∏–≤–æ, —É –≤–∞—Å —î –∑–¥–æ–≥–∞–¥–∫–∏, —á–æ–º—É `from math import sin as m_sin, cos as m_cos, exp as m_exp` –≤–∏–Ω–µ—Å–µ–Ω–æ –æ–∫—Ä–µ–º–æ—é –∫–æ–º—ñ—Ä–∫–æ—é? 
–Ü —á–æ–º—É –º–∏ –∫–æ—Ä–∏—Å—Ç—É—î–º–æ—Å—å `m_sin` –∑–∞–º—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ–± –ø–∏—Å–∞—Ç–∏ `math.sin`? 

--------------------

–°–ø–æ–¥—ñ–≤–∞—é—Å—å, –∑ –ø–µ—Ä–µ–∫–ª–∞–¥–æ–º —ñ–Ω—à–∏—Ö —Ñ–æ—Ä–º—É–ª –∑ –º–∞—Ç–µ–º–∞—Ç–∏—á–Ω–∏—Ö –Ω–æ—Ç–∞—Ü—ñ–π –Ω–∞ "–º–æ–≤—É NumPy" –ø—Ä–æ–±–ª–µ–º –Ω–µ –≤–∏–Ω–∏–∫–Ω–µ.

–î–∞–≤–∞–π—Ç–µ –ø–æ–≥–ª—è–Ω–µ–º–æ, —â–æ —â–µ —î –≤ NumPy. 
`numpy.linalg` - –ø–æ—Ç—É–∂–Ω–∏–π –º–æ–¥—É–ª—å, —è–∫–∏–π —Ä–æ–∑–≤ º—è–∑—É—î –æ—Å–Ω–æ–≤–Ω—ñ –ø—ñ–¥–∑–∞–¥–∞—á—ñ –ª—ñ–Ω—ñ–π–Ω–æ—ó –∞–ª–≥–µ–±—Ä–∏.

–î–∞–≤–∞–π—Ç–µ, –¥–æ –ø—Ä–∏–∫–ª–∞–¥—É, –∑–Ω–∞–π–¥–µ–º–æ –º–∞—Ç—Ä–∏—Ü—é, –æ–±–µ—Ä–Ω–µ–Ω—É –¥–æ:
$$
\begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 10
\end{pmatrix}
$$

–©–æ –∂, —Å–ø–µ—Ä—à—É –¥—ñ–∑–Ω–∞—î–º–æ—Å—å, —á–∏ —Ü–µ –≤–∑–∞–≥–∞–ª—ñ –º–æ–∂–ª–∏–≤–æ. –Ø–∫–∏–π –∫—Ä–∏—Ç–µ—Ä—ñ–π –æ–±–µ—Ä—Ç–Ω–æ–≤–Ω–æ—Å—Ç—ñ –º–∞—Ç—Ä–∏—Ü—ñ? –ó–¥–æ–≥–∞–¥–∞—î—Ç–µ—Å—å, —è–∫–∞ —Ñ—É–Ω–∫—Ü—ñ—è —É `np.linalg` –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î –∑–∞ —Ü–µ?


````python
matrix = np.array([[1, 2, 3], [4, 5, 6], [7, 8, 10]])
````

````python
DET = np.linalg.det(matrix)
DET != 0
````

–¢–µ–ø–µ—Ä, —è–∫—â–æ –º–∏ –≤–∂–µ –∑–Ω–∞—î–º–æ, —â–æ –º–∞—Ç—Ä–∏—Ü—è –æ–±–µ—Ä—Ç–æ–≤–Ω–∞, –¥–∞–≤–∞–π—Ç–µ —Ç–∞–∫–∏ –∑–Ω–∞–π–¥–µ–º–æ –æ–±–µ—Ä–Ω–µ–Ω—É –¥–æ –Ω–µ—ó. –í–∞–º –∑–Ω–∞–¥–æ–±–∏—Ç—å—Å—è —Ñ—É–Ω–∫—Ü—ñ—è `np.linalg.inv`.

–ó–∞–ø–∏—à—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é —É –∑–º—ñ–Ω–Ω—É `INV`

````python
INV = np.linalg.inv(matrix)
INV
````

–¢–µ–ø–µ—Ä –¥–∞–≤–∞–π—Ç–µ —Ç—Ä–æ—Ö–∏ –ø–æ–≥—Ä–∞—î–º–æ—Å—è –∑ –º–∞—Ç—Ä–∏—Ü–µ—é `INV`. –ü–µ—Ä—à –∑–∞ –≤—Å–µ, —Å–ø—Ä–æ–±—É—î–º–æ –æ—Ç—Ä–∏–º–∞—Ç–∏ —É—Å—ñ –µ–ª–µ–º–µ–Ω—Ç–∏, —è–∫—ñ –º–µ–Ω—à—ñ –∑–∞ 0.5, —É –≤–∏–≥–ª—è–¥—ñ –≤–µ–∫—Ç–æ—Ä–∞:

````python
sel_inv = INV[INV < 0.5]
sel_inv
````

–°—É–ø–µ—Ä! –¢–µ–ø–µ—Ä –¥–∞–≤–∞–π—Ç–µ —Å–ø—Ä–æ–±—É—î–º–æ –¥–æ–º–Ω–æ–∂–∏—Ç–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω—ñ –µ–ª–µ–º–µ–Ω—Ç–∏ –Ω–∞ 2. –ù–∞—à–∞ –º–µ—Ç–∞ - —â–æ–± —Å–∞–º–∞ –º–∞—Ç—Ä–∏—Ü—è `INV` –∑–º—ñ–Ω–∏–ª–∞—Å—è, –∞ –Ω–µ –≤–∞—à –Ω–æ–≤–∏–π –≤–µ–∫—Ç–æ—Ä.

````python
INV[INV < 0.5] *= 2
INV
````

–ö—Ä—É—Ç–æ! 
–¢–µ–ø–µ—Ä –¥–∞–≤–∞–π—Ç–µ –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞, —è–∫–∏–π –º–µ–Ω—à–∏–π –∑–∞ 0.3, —Å–ø—Ä–æ–±—É—î–º–æ –∑–∞—Å—Ç–æ—Å—É–≤–∞—Ç–∏ –Ω–∞—Å—Ç—É–ø–Ω—É —Ñ—É–Ω–∫—Ü—ñ—é:
$$ f(x) = x^2 - sin(x) $$

–¢—É—Ç —î –¥–µ–∫—ñ–ª—å–∫–∞ –≤–∞—Ä—ñ–∞–Ω—Ç—ñ–≤, –∞–ª–µ —Å–ø—Ä–æ–±—É–π—Ç–µ –≤ –ø–µ—Ä—à—É —á–µ—Ä–≥—É —Ä–æ–∑—ñ–±—Ä–∞—Ç–∏—Å—è –∑ —Ç–∏–º, —è–∫ –ø—Ä–∞—Ü—é—î `vectorize`

````python
def f(x):
    return np.pow(x, 2) - np.sin(x)

vectorized_f = np.vectorize(f)
boolean_mask = INV < 0.3

INV[boolean_mask] = vectorized_f(INV[boolean_mask])
INV
````

–ú–∞—Ç—Ä–∏—Ü—ñ —Ç–æ –¥–æ–±—Ä–µ, –∞–ª–µ NumPy –º–æ–∂–µ —Å—Ç–∞—Ç–∏ —É –Ω–∞–≥–æ–¥—ñ –Ω–∞–≤—ñ—Ç—å —à–∫–æ–ª—è—Ä–µ–≤—ñ. –ü–æ–≥–ª—è–Ω—å—Ç–µ, —â–æ –∑ —Å–µ–±–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î —Ñ—É–Ω–∫—Ü—ñ—è np.linalg.solve. –ü—ñ—Å–ª—è —Ü—å–æ–≥–æ —Å–ø—Ä–æ–±—É–π—Ç–µ –≤–∏—Ä—ñ—à–∏—Ç–∏ —Å–∏—Å—Ç–µ–º—É –ª—ñ–Ω—ñ–π–Ω–∏—Ö —Ä—ñ–≤–Ω—è–Ω—å:
$$
\begin{align*}
x + 4y - 2z &= 1 \\
2x + 4z &= 2 \\
-3x + 2y + 3z &= 3
\end{align*}
$$


````python
A = np.array([[1, 4, -2],
              [2, 0, 4],
              [-3, 2, 3]])

b = np.array([1, 2, 3])

x = np.linalg.solve(A, b)
x
````

### –ü–û–ß–ê–¢–û–ö –î–û–ú–ê–®–ù–¨–û–á –ß–ê–°–¢–ò–ù–ò

    –ú–∏ –≤–∂–µ —Ç—Ä–æ—Ö–∏ –ø–æ–≥–ª—è–Ω—É–ª–∏, –Ω–∞—Å–∫—ñ–ª—å–∫–∏ NumPy —à–≤–∏–¥—à–µ, –Ω—ñ–∂ —Å–∞–º–æ–ø–∏—Å–Ω–∏–π –∫–æ–¥. –ê–ª–µ —ñ–Ω–æ–¥—ñ –≤—ñ–Ω –±—É–≤–∞—î –Ω–µ–ø—Ä–∏—Å—Ç–æ–π–Ω–æ —à–≤–∏–¥—à–µ, –æ—Å–æ–±–ª–∏–≤–æ –ø–æ—Ä—ñ–≤–Ω—é—é—á–∏ –∑ –∫–æ–¥–æ–º –ø–æ–≤–Ω—ñ—Å—Ç—é –Ω–∞ Python.

    –í–∞—à–µ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞—Å—Ç—É–ø–Ω–µ: —Å–ø—Ä–æ–±—É–π—Ç–µ –Ω–∞–ø–∏—Å–∞—Ç–∏ –∑–≤–∏—á–∞–π–Ω–∏–π –∞–ª–≥–æ—Ä–∏—Ç–º –¥–ª—è –º–Ω–æ–∂–µ–Ω–Ω—è –¥–≤–æ—Ö –º–∞—Ç—Ä–∏—Ü—å (–≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ª–∏—à–µ –∑–≤–∏—á–∞–π–Ω–∏–π list —Ç–∞ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü—ñ—ó Python, –±–µ–∑ NumPy) —ñ –ø—Ä–æ—Ç–µ—Å—Ç—É–π—Ç–µ –π–æ–≥–æ –Ω–∞ –º–∞—Ç—Ä–∏—Ü—è—Ö —Ä–æ–∑–º—ñ—Ä—É 200x200. –ü–æ—Ä—ñ–≤–Ω—è–π—Ç–µ —Ü–µ –∑—ñ —à–≤–∏–¥–∫—ñ—Å—Ç—é NumPy. –ó—Ä–æ–±—ñ—Ç—å –≤–∏—Å–Ω–æ–≤–∫–∏ :)

    –ú–∞—Ç—Ä–∏—Ü—ñ, —è–∫—ñ —Ç—Ä–µ–±–∞ –ø–æ–º–Ω–æ–∂–∏—Ç–∏, –∞ —Ç–∞–∫–æ–∂ –º–∞—Ç—Ä–∏—Ü—è –¥–ª—è –∑–∞–ø–∏—Å—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É –ª—é–± º—è–∑–Ω–æ —Å—Ç–≤–æ—Ä–µ–Ω—ñ –Ω–∏–∂—á–µ


````python


M1 = [[i] * 200 for i in range(200)]
M2 = [[i + 1] * 200 for i in range(200)]

M_RES = [[0] * 200 for i in range(200)]
````

````python

for i in range (len(M1)):
    for j in range(len(M1[0])):
        for k in range(200):
            M_RES[i][j] += M1[i][k] * M2[k][j]

print(M_RES)
````

````python

M1_np = np.array(M1)
M2_np = np.array(M2)

M_RES_np = M1_np @ M2_np
M_RES_np

````
